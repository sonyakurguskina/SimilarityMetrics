```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#спарсим статью из википедии через апи википедии, я решила проверить статью про науку
wikipedia.set_lang("ru")
wiki_page = wikipedia.page("Наука")
page_content = wiki_page.content
```

```python
with open("content.txt", "w",  encoding='utf-8') as file:
    file.write(page_content)
```

```python
def words(text):               #метод для обработки текста, очищаем от знаков препинания и разделяем по словам
    words = []
    with open(text, "r", encoding="utf-8") as file: 
        for line in file.readlines():
            string = re.sub("[^\w\s]", "", line)
            words_in_line = string.split()
            words = words + words_in_line
        return words
```

```python
wiki_words = words("content.txt") #разбиваем статью на слова
```

```python
#приводим все слова в нормальную форму
morph = pymorphy2.MorphAnalyzer()
word_list = []

for i in range(len(wiki_words)):
    p = morph.parse(wiki_words[i])[0]  
    word_list.append(p.normal_form)

print(word_list)
```

```python
#очищаем статью от стоп слов
remove_sw = [word for word in word_list if not word in stopwords.words()]
print(remove_sw)
```

```python
#файлы с ключевыми словами, разбиваем их по запятой
science = words("наука.txt")
sport = words("спорт.txt")
shopping = words("шоппинг.txt")
news = words("новости.txt")
```

```python
#метрики совпадения ключевых слов с исходным текстом(учитывая стоп-слова),
#здесь рассчитан процент нахождения ключевых слов по категориям в статье
science_sim = len([word for word in remove_sw if word in science])/len(word_list)        # 16%
sport_sim = len([word for word in remove_sw if word in sport])/len(word_list)            # 2%
shopping_sim = len([word for word in remove_sw if word in shopping])/len(word_list)      # 0,5%
news_sim = len([word for word in remove_sw if word in news])/len(word_list)              # 3%
print(science_sim,sport_sim,shopping_sim,news_sim)
```

```python
def jac(x: set, y: set):          #метод для метрики Жаккарда
    shared = x.intersection(y) 
    return len(shared) / len(x.union(y)) 
```

```python
print(jac(set(remove_sw), set(science)))
print(jac(set(remove_sw), set(sport)))
print(jac(set(remove_sw), set(shopping)))
print(jac(set(remove_sw), set(news)))
```

```python
#метод для метрики Косинуса
def cos(x,y):
    cleaned = [str(x), str(y)]
    vectorizer = CountVectorizer().fit_transform(cleaned)
    vectors = vectorizer.toarray()
    return cosine_similarity(vectors[0:1], vectors)
```

```python
print(cos(remove_sw, science))
print(cos(remove_sw, sport))
print(cos(remove_sw, shopping))
print(cos(remove_sw, news))
```
